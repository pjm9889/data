### <div class="alert alert-block alert-info">  [Part1] 개요 </div>
- 수집된 자료의 통계적 특성을 파악하기 위해서는 각종 그래프를 이용(시각화) 하며 정확하게 수치로 분석하여 표시하는 척도가 필요하다.
- 이러한 수치로서 통계적 정보를 나타낸 것을 통계치 또는 통계량(statistic)라 한다.
<br><br>
- 가장기본적인 통계량으로는 데이터의 특성이나 양적 구조의 특성이 무엇이며 모집단 분포의 중심으로서 전체를 대표하는 값이 어떤 것인가를 나타내는 즉, 데이터의
 집중화 경향을 나타내어주는 대표치가 있다.
    -  집중화경향: 자료의 중심을 측정하는 척도 (mean, median, mode 최빈값, mid-range 범위의중앙값)
    <br><br>
    
- 또한 이들 데이터들이 대표치의 주위에 어느정도 흩어져 있는지를 계량적으로 측정하는 척도가 필요하게 되는데 이들 데이터의 흩어진 정도를 나타내어 주는것을 산포도(dispersion)라 한다.
    - 산포도: 관측값의 흩어진 정도를 나타내는 척도(range 범위, variance 분산, standard deviation 표준편차, coefficient of variation 변동계수: 측정단위가 서로 다른 자료를 비교하고자 할 때 쓰임-->표준 편차 / 평균) 등이 있다.
    <br><br>
- 또한 데이터가 분포하고 있는 모양을 비교하여주는 통계량으로 비대칭도(왜도: skewness)와 첨도(kurtoisis)등이 있다.
#     - 비대칭도: 좌우대칭의 분포를 기준으로 분포가 왼쪽이나 오른쪽으로 치우쳐 있을때 그 정도를 나타내는 척도이다.
     - 첨도는 분포의 모양이 얼마나 뾰족한가를 나타내는 척도로서 보통 정규분포를 기준으로 하여 상대적으로 높은 첨도, 낮은 첨도로 구분하게 된다.

### <div class="alert alert-block alert-info">  [Part4] 왜도와 첨도 </div>
-  왜도: 정규분포로부터 얼마자 벗어나 있는가의 정도를 나타내는 지표로 사용하며 피어슨의 비대칭 계수를 사용할수 있음.
  
    - 분포가 좌우대칭일때의 왜도는 0
    - 왜도가 양수이면 분포의 비대칭 꼬리가 양의 값 쪽으로 길게 늘어져 있으며, 왜도가 음수이면 분포의 비대칭 꼬리가 음의 값쪽으로 길게 늘어져 있음
        <br><br>
- 피어슨비대칭계수: 3(평균-중앙값)/표준편차
    - 비대칭 계수의 0: 평균과 중앙값의 크기가 같음 (좌우대칭분포)
    - 비대칭 계수 음수: 평균이 중앙값보다 작음 (오른쪽 꼬리 비대칭 분포)
    - 비대칭 계수 양수: 평균이 중앙값보다 큼 (왼쪽 꼬리 비대칭 분포)

### 5-1. 표본의 크기가 작은경우
- 모집단의 분산이 알려지지 않고 또한 표본의 크기가 작은경우:
    - 모평균에 관한 구간추정은 모집단이 정규분포를 따르는 경우에 가능하다.
    - 모집단이 알려지지 않았음으로 표본분산을 사용하며, 자유도(n-1)인 t분포 를 따른다.
    - t-분포는 표준정규분포와 같이 0을 중심으로 종모양의 대칭 형태를 갖는다.
        - 표준정규분포: 표준정규분포(Standard Normal Distribution)는 정규분포를 표준화(Standardization)하여 얻어진 확률분포
    - 그러나 t-분포는 정규분포에 비교할때 더 두터운 꼬리를 갖는다는 차이점이 있으나 자유도가 커질수록 t-분포의 형태는 정규분포의 형태로 가까워진다.

```python
np.random.seed(0)

rv = sp.stats.norm()
M = 1000

plt.subplot(1, 2, 1)
N = 4
x1 = rv.rvs((N, M))
xbar1 = x1.mean(axis=0)
xstd1 = x1.std(axis=0, ddof=1)
x = xbar1 / (xstd1 / np.sqrt(N))
sns.distplot(x, kde=True)
xx = np.linspace(-6, 6, 1000)
plt.plot(xx, rv.pdf(xx), 'r:', label="정규분포")
plt.xlim(-6, 6)
plt.ylim(0, 0.5)
plt.title("t 통계량 분포 (N = 4)")
plt.legend()

plt.subplot(1, 2, 2)
N = 40
x2 = rv.rvs((N, M))
xbar2 = x2.mean(axis=0)
xstd2 = x2.std(axis=0, ddof=1)
x = xbar2 / (xstd2 / np.sqrt(N))
sns.distplot(x, kde=True)
xx = np.linspace(-6, 6, 1000)
plt.plot(xx, rv.pdf(xx), 'r:', label="정규분포")
plt.xlim(-6, 6)
plt.ylim(0, 0.5)
plt.title("t 통계량 분포 (N = 40)")
plt.legend()

plt.show()
```

### <div class="alert alert-block alert-info">  [Part6] 가설검정 </div>
- 모집단으로부터 얻어진 표본정보를 이용하여 모수에 대한 추정을 한 이후에는 모수가 갖으리라고 예측한 값이 올바르게 얻어졌는지를 판정하는 과정이 필요하다. 이러한 과정을 가설검정이라함.
    - 1) 모집단의모수에 대한 가설(hypothesis)을 세우고
    - 2) 기각역, 검정통계량을 계산하여
    - 3) 결과판정의 과정을 거침
<br><br>
- 예) A 전자회사의 경영진은 자사의 제품으 구입한후 다시 자사의 제품을 재구매 할 소비자의 비율을 70%로 파악하고 있다. A제품의 소비자중 500명의 소비자를 대상으로 새롭게 제품의 여부를 조사하였더니 재구매를 응답한 비율은 60%로 조사되었다.
    -> 실제 재구매 비율은 10% 하락하였는가?
    

### 6-1. 귀무가설과 대립가설
- 귀무가설: H0(영가설)--> 모수에 관하여 이미 알려진 사실을 사용하여 세워지는 가설
- 대립가설: H1--> 귀무가설이 옳지 않은 경우 받아들이는 가설
<br><br>
- 가설검정의 근본적인 관심은 제기되는 문제를 반영하는 대립가설의 타당성 여부를 확인하는 것임.
- 그러나 직접적으로 대립가설만을 사용하여 대립가설의 타당성 여부를 검정하지 반대되는 귀무가설을 세우게 되며 귀무가설의 옳지 않은경우에 대립가설의 타당성응ㄹ 받아들이는 과정으로 가설검정이 진행된다.

    - 예) 귀무가설: 재구매할 소비자의 비율은 70%
           -대립가설: 재구매를 70%가아닌 60%이다 임으로
           
         
    
 <br><br>

### 6-2. 가설검증 오류
 - 1종오류: 귀무가설(H0)이 참인데 --> 귀무가설을 기각하고 대립가설(H1)을 선택
 - 2종오류: 귀무가설(H0)이 거짓인데 --> 귀무가설을 선택
 - 1종과 2종 오류를 동시에 줄일수는 없다.
 - 일반적으로 1종오류가 2종오류보다 더욱 중요한 오류로서 고려되어 1종 오류의 발생 확률을 제한하며 이러한 확률을 유의수준이라고 한다. 유의수준의값으로는 0.01, 0.05, 0.1이 널리 사용된다.

 <br>

 ### 6-3. 검정통계량(test statistic)
 - 귀무가설과 대립가설중에서 하나의 가설을 입증하기 위하여 사용하는 통계량
     - 귀무가설이 옳다는 가정하에서 검정통계량의 분포를 구함
     - 예) 관심모수가 모평균 또는 모비율이라면 표본평균을 표준화한 통계량 또는 표본비율을 표준화한 통계량을 각각의 검정통계량으로 설정

P-value

관찰된 데이터가 귀무가설과 양립하는 정도를 0~1 사이의 수치로 표현한 것.
p-value값이 낮다는 것은, 귀무가설이 틀렸을 확률이 높다.

- 1) pvalue < 0.01
: 귀무가설이 옳을 확률이 1%이하 -> 틀렸다 (깐깐한 기준)
- 2) pvalue < 0.05 (5%)
: 귀무가설이 옳을 확률이 5%이하 -> 틀렸다 (일반적인 기준)
- 2.5) 0.05 ~ pvalue ~ 0.1 사이인 경우: (애매함)
실험을 다시한다. / 데이터를 다시 뽑는다. / 샘플링을 다시한다기존의 경험 / 인사이트를 바탕으로 가설에 대한 결론을 내린다.

- 3) pvalue > 0.1 (10%)
: 귀무가설이 옳을 확률이 10%이상 -> 귀무가설이 맞다 ~ 틀리지 않았을것이다 사이
