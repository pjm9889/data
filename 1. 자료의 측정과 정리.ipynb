### <div class="alert alert-block alert-info">  [Part2] 자료의 종류 </div>
- 질적연구: 연구문제를 이해하는 목적으로 적은수의 대표성이 없는 표본을 가지고 진행하는 바법으로 구조화 되지 않은 탐색적 연구방법이다. (통계분석을 사용하지 않는 범주형 자료)
    - 직업, 거주지역, 결혼여부등과 같은 질적 특성에 근거하여 분류하는 관측값으로 기술통계가 아닌 갯수를 구하는 산술식만 사용함
    <br><br>
- 양적연구: 자료수집방법이 구조화되어 있으며 대표성을 갖는 많은수의 표본을 대상으로 발견된 사실을 일반적인 상황으로 확장하는 연구방법이다. (통계분석을 사용하는 수치적 자료)
    - 연속적자료(키, 몸무게와 같은 실수형자료로서 모든 값을 취할수 있음) 
    - 이산적자료(자녀수 처럼 셀수있는 숫자로만 되어 있는 정수형자료)
    <br><br>
- 집단화(grouping): 양적자료는 필요에 따라 질적자료로 변환(예: 이산형 나이값을 20-29의 20대, 30대와 같은 질적자료로 변환)
- (참고: 경영/경제에서 이용되는 자료)
    - 횡단면자료: 연구대상을 일정시점에서 각 주체별로 관측하여 얻은값 --> 같은시간에 팔린 아메리카노와 케익의수
    - 시계열자료: 연구대상이 되는 어떤 특정변수들을 시간에 따라 일정기간동안 관측하여 얻은 값을 말함 --> 일자별 판매금액
    
<div class="alert alert-block alert-warning">  성별, 직업, 만족도, 구매의사, 선호회사는 질적 자료에 속하며 나이와 월수입은 양적자료에 속한다. 또한 나이와 월수입은 원칙적으로는 연속적 자료에 속하는 것이나 조사 단위에 의하여 이산적 자료로 간주할수 있다. </div>

### <div class="alert alert-block alert-info"> [Part3] 자료의 측정 </div>
- 측정(measurement): 미리 결정된 규칙에 의하여 연구대상의 특징을 수치나 기호로 할당하는 과정
    - 연구대상의 특성이나 태도에 대한 특징이 어떠한가에 초점을 두는것이다.
        -예: 고객대상마케팅조사: 고객이 지작하고 있는 서비스품질이나, 선호도등의 특성을 수치나 기호로 할당하는 규칙을 정하는것
    - 하나의 특성을 여러 가지 수치나 기호로 표현한다면 측정이 불가능해질 것이므로 연구대상의 특성을 일관성이 있도록 적용할수 있어야 함. 따라서 연구대상의 특성을 측정하는데 있어서는 척도를 만들어 사용하는데 실제 관측값을 어떻게 수치와 대응시키느냐에 따라 여거 가지 척도를 생각할수 있다.
        - 예:학교내 식당에 대한 서비스품질에 대한 조사실시-->특성을 측정하기 위한 설문--> 특성항목을 응답자들이 쉽게 답할수 있도록 관측값들을 수치와 대응시키도록 해야함.(즉, 식당내 음식의 맛을 기호로 표현해야함)
            - 1: 매우 맛없다.
            - 2: 조금 맛없다
            - 3: 보통이다.
            - 4: 조금 맛없다.
            - 5: 매우 맛없다.
            - 이렇게 1-5까지 관측값들에 기호를 부여하게 되는데 가장 널리 이용되고 있는척도에는 명목척도와 서열척도, 구간척도, 비율척도가 있다.

#### 3-1. 명목척도(nominal scale)
- 명목: 사물의 이름, 호칭의 의미
- 명목척도에 이용가능한 수치는 각 범주에 속하는 자료의 수밖에 없음으로 집중화 경향(tendency)를 측정하기 위해서는 최빈값(mode)이 이용되며, 통계적 유의성을 검증하는 방법에는 X2검정(chi_square test:카이제곱,https://blog.naver.com/parksehoon1971/221589203965), 이항검정(binomial test)등이 있다. 


#### 3-2. 서열척도(ordinal scale)
- 연구대상의 특성에 순성가 있음. 서열척도에서의 수치는 서열로서의 의미밖에 없기 때문에 집중화 경향치로서는 중앙치(median)이 이용되며 산포도를 측정하기 위해서는 백분위수나 사분위수가 이용된다. 
    - 통계적 유의성을 검증하는데에는 비모수통계방법(nonparametric method)만 사용가능함
            - 순위상관계수(rank-order correlation)
        - 순위를 이용한 분산분석(Friedman's ANOVA)등이 있음
-*비모수통계방법: 관측 값이 어느 특정한 확률분포를 따른다고 전제할 수 없거나 또는 모집단에 대한 아무런 정보가 없는 경우에 실시하는 검정방법으로 모수에 대한 언급 이 없으며 분포무관 방법이라고도험
- 

#### 3-3. 구간척도(interval scalee)=등간척도, 간격척도
- 각 부분간의 간격이 같다는 의미로서 서열의 개념과, 거리 및 간격의 개념을 포함하고 있다. 
     - 예) 시간과 온도: 1시부터 3시까지와 3시부터 6시까지으 시간간격은 동일하다.
- 학교식당의 1,2,3,4,5점은 구간척도에 해당함.
- 구간척동의 집중화 경향치로는 산술평균(mean)이 이용됨
- 산포도를 알기위하여서는 분산또는 표준편차가 이용되며, 통계적 기법으로는 t-검정, F-검정 및 분산분석, 회귀분석, 피어슨상관계수등을 이용할수 있으나, 기하평균, 조화평균, 변동계수와 같은 특별한 통계량은 등간척도에 사용할수 없다.
<br><br>
    - 기하평균
    - 조화평균
    - 변동계수
    

#### 3-4. 비율척도(ratio scale)
- 명목, 서열, 구간 척도의 특징을 모두 포함하고 있으며 여기에 원점(origin point) 또는 영점(zero point)의 개념이 부가된것이다.
    - (예) 키와 몸무게, 거리, 면적과 같은 수치에 원점이 존재하기 때문에 비율(%)로 계산이 가능ㅎ하다.
    - 비율척도는 모수적통계방법이 사용될수 있다. 산포도를 알기위하여 변동계수가 이용될수 있으며 집중화 경향치로 기하평균과 조화평균이 이용될수 있다.
